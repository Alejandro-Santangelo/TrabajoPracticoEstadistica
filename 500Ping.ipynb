```python
import pandas as pd
ValoresPingBna=pd.read_excel("/content/drive/MyDrive/NARCISO/ValoresPingBna.xlsx")
ValoresPingBna
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Graficando histograma
mu, sigma = 0, 0.2 # media y desvio estandar
datos = np.random.normal(mu, sigma, 1000) #creando muestra de datos

# histograma de distribución normal.
cuenta, cajas, ignorar = plt.hist(datos, 20)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma')
plt.show()
```

```python
from scipy.stats import norm

# Calcular la media y la desviación estándar de los datos
media = ValoresPingBna['time'].mean()
desviacion_estandar = ValoresPingBna['time'].std()

# Crear un rango de valores para el eje x
x = np.linspace(ValoresPingBna['time'].min(), ValoresPingBna['time'].max(), 500)

# Calcular la función de densidad de probabilidad normal (campana de Gauss)
y = norm.pdf(x, media, desviacion_estandar)

# Crear el gráfico
plt.plot(x, y, color='blue')
plt.title('Campana de Gauss')
plt.xlabel('Valores')
plt.ylabel('Densidad de probabilidad')
plt.axvline(media, color='red', linestyle='--', label='Media')  # Agregar línea vertical en el valor medio
plt.grid(True)
plt.show()
```
